```python
# انتخاب ستون‌های عددی
import numpy as np

numeric_df = df.select_dtypes(include=['number'])

# نمایش ماتریس همبستگی
corr_matrix = numeric_df.corr()
print("Correlation matrix:\n", corr_matrix)

# مرتب‌سازی زوج‌های همبستگی به صورت تک بعدی (بدون تکرار)
corr_unstacked = corr_matrix.where(~np.eye(corr_matrix.shape[0],dtype=bool)).unstack().dropna()

# چاپ 5 جفت متغیر با بیشترین همبستگی مثبت
print("\nTop 5 positive correlations:")
print(corr_unstacked.sort_values(ascending=False).head(5))

# چاپ 5 جفت متغیر با بیشترین همبستگی منفی
print("\nTop 5 negative correlations:")
print(corr_unstacked.sort_values().head(5))
```

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Numeric Columns')
plt.show()
```

تحلیل ماتریس همبستگی ستون‌های عددی



ماتریس همبستگی نشان می‌دهد که بیشترین ارتباط مثبت قوی بین تعداد تماس‌های قبلی (previous) و تعداد روزهای گذشته از آخرین تماس (pdays) با ضریب حدود 0.45 است. این ارتباط منطقی است؛ هرچه تماس‌های قبلی بیشتر باشد، فاصله کمتری از آخرین تماس وجود دارد.

همچنین، همبستگی ضعیف مثبت بین سن مشتری (age) و موجودی حساب بانکی (balance) دیده می‌شود که نشان می‌دهد سن و موجودی ارتباط قابل توجهی ندارند.

متغیرهای دیگری مانند تعداد تماس‌های جاری (campaign) و روز تماس (day) هم همبستگی مثبت کمی دارند که می‌تواند نشان‌دهنده الگوهای تماس در روزهای خاص باشد.

بیشترین همبستگی منفی ضعیف بین pdays و day مشاهده می‌شود که می‌تواند نشان‌دهنده پراکندگی زمانی تماس‌ها باشد.


تمرکز روی متغیرهای previous و pdays در تحلیل‌های بعدی می‌تواند به بهبود مدل پیش‌بینی کمک کند. همچنین، توجه به الگوهای زمانی تماس‌ها (day و campaign) برای فهم بهتر رفتار مشتریان ضروری است.

```python
# گروه‌بندی داده‌ها بر اساس نتیجه کمپین
grouped = df.groupby('y')

# نمایش آمار توصیفی ستون age و balance برای هر گروه (yes و no)
desc = grouped[['age', 'balance']].describe()
print(desc)

# تعداد نمونه‌ها در هر گروه
counts = grouped.size()
print("\nNumber of samples in each group:\n", counts)

# مشاهده نقاط پرت (outliers) با نمایش بیشینه و کمینه
max_values = grouped[['age', 'balance']].max()
min_values = grouped[['age', 'balance']].min()
print("\nMax values:\n", max_values)
print("\nMin values:\n", min_values)
```
