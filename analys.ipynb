```python
for col in ['job', 'education']:
    total = len(df)
    unknown_count = df[df[col] == 'unknown'].shape[0]
    unknown_percent = (unknown_count / total) * 100
    print(f"Column '{col}':")
    print(f"  Unknown count = {unknown_count}")
    print(f"  Unknown percentage = {unknown_percent:.2f}%\n")
```

```python
df_clean = df[(df['job'] != 'unknown') & (df['education'] != 'unknown')].copy()
print(f"Original dataset size: {len(df)}")
print(f"Cleaned dataset size after removing 'unknown': {len(df_clean)}")
```

```python
df_clean = df[(df['job'] != 'unknown') & (df['education'] != 'unknown')]

sns.countplot(data=df_clean, x='job', palette='Set3')
plt.xticks(rotation=45)
plt.title('Jobs (Unknown Removed)')
plt.show()

sns.countplot(data=df_clean, x='education', palette='Set2')
plt.xticks(rotation=45)
plt.title('Education (Unknown Removed)')
plt.show()
```

```python
#   
import numpy as np

numeric_df = df.select_dtypes(include=['number'])

#   
corr_matrix = numeric_df.corr()
print("Correlation matrix:\n", corr_matrix)

#        ( )
corr_unstacked = corr_matrix.where(~np.eye(corr_matrix.shape[0],dtype=bool)).unstack().dropna()

#  5      
print("\nTop 5 positive correlations:")
print(corr_unstacked.sort_values(ascending=False).head(5))

#  5      
print("\nTop 5 negative correlations:")
print(corr_unstacked.sort_values().head(5))
```

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Numeric Columns')
plt.show()
```
